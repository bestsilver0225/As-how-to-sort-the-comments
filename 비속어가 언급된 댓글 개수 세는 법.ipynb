```python
import pandas as pd
import re
```

```python
#데이터불러오기
```

```python
data_005=pd.read_csv('005.csv')
```

```python
data_005=data_005[['text','file','vector']]#데이터에서 원하는 열만 남기기
```

# 2. 댓글에서 특수기호 제거

```python
data_005['text']=data_005['text'].apply(lambda x : re.sub(r'[^\w\s]', '', x)) # 정규표현식을 사용하여 특수기호를 제거
```

```python
data_005
```

# 3. 비속어 세기

```python
f = open('욕.txt', 'r',encoding = "utf-8")# 비속어가 담긴 파일 읽어오기
ff = f.readlines()


def how_many(text):
    for i in ff: 
            text=text.replace(i.strip(),"최고은",1) # 댓글에 포함된 비속어를 제 이름으로 바꿨어요
    return text.count("최고은") # 그리고 제 이름이 몇번 나오는지 카운트하는 함수를 만들었습니다
            
data_005['단순횟수세기']=data_005['text'].apply(how_many) # 그리고 제가 만든 함수를 각 댓글에 적용해 보았습니다.
```

```python
data_005
```

# 4. 각 기사의 전체 댓글 중  비속어가 언급된 댓글 개수 비율 구하기

```python
all=data_005.groupby('file').count()[['text']] # 각 기사 (file)에 댓글이 몇개 달렸는지 세어보았습니다.
all
```

```python
all=all.reset_index() # 인덱스를 열로 바꿔줄게요
```

```python
all
```

```python
set(data_005['단순횟수세기']) # 욕설이 전혀 언급되지 않은 댓글도 있고 욕설이 7번 언급된 댓글도 있네요
```

```python
def one_or_zero(num): # 욕설이 한번도 언급되지 않으면 욕설이 여러번 언급되었으면 1로 반환하는 함수에요
    if num == 0:
        return 0
    else:
        return 1
```

```python
data_005['욕설유무']=data_005['단순횟수세기'].apply(one_or_zero) # 그리고 제가 만든 함수를 각 댓글에 적용해 보았습니다.
```

```python
data_005
```

```python
yes_or_not=data_005.groupby('file').sum()[['욕설유무']] # 각 기사당 욕설이 포함된 댓글의 개수 구하기
yes_or_not
```

```python
yes_or_not=yes_or_not.reset_index()# 인덱스를 열로 바꿔줄게요
```

```python
yes_or_not
```

```python
final=pd.merge(all,yes_or_not,on='file') # 위에서 구한 두 데이터 프레임을 합쳐요
final
```

```python
final['욕설이 있는 댓글 비율']=final['욕설유무']/final['text']*100
final
```

```python
final.to_csv('욕설_005_비율_final.csv') # 파일로 저장해요
```
